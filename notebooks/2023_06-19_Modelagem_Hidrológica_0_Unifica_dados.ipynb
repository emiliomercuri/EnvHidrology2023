* Evapotranspiração MODIS
* Precipitação CHIRPS
* Vazão ANA

```python
formato_data = lambda x: pd.datetime.strptime(x, "%d/%m/%Y")
df_P_chirps = pd.read_csv('ClimateEngine_dados_CHIRPS_CAMBARA2.csv', index_col=0, sep=';', date_parser=formato_data)
df_P_chirps.index = pd.to_datetime(df_P_chirps.index)
df_P_chirps
```

```python
formato_data = lambda x: pd.datetime.strptime(x, "%d/%m/%Y")
df_ET_modis = pd.read_csv('evap_MODIS.csv', index_col=0, sep=';', date_parser=formato_data)
df_ET_modis.index = pd.to_datetime(df_ET_modis.index)
df_ET_modis
```

```python
formato_data = lambda x: pd.datetime.strptime(x, "%d/%m/%Y")
df_vazao = pd.read_csv('vazao-82220000.csv', index_col=0, sep=';', date_parser=formato_data)
df_vazao.index = pd.to_datetime(df_vazao.index)
df_vazao
```

```python
df_P_chirps['2004':'2008']
```

```python
df_geral = df_P_chirps['2004':'2008']
df_geral['prec(mm/dia)'] = df_geral['(mm)Precipitation(CHIRPS)']
df_geral = df_geral[['prec(mm/dia)']]
df_geral['etp(mm/dia)'] = df_ET_modis['PET(mm/d)']
df_geral['et(mm/dia)'] = df_ET_modis['ET(mm/d)']
df_geral['q(m3/s)'] = df_vazao
df_geral
```

```python
df_geral.to_csv('dados_cambara_2004-2008.csv')
```
